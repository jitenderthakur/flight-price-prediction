## Finding and Removing Multicollinearity from data

```python
threshold=0.9
```

```python
# find and remove correlated independent features
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) >= threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
# printing independent variable which are highly correlated with each other
correlation(X,threshold)
```

#### From the above results we can infer that the many features are having strong multicollinearity in the data set.

```python
# droping column which having multicollinearity >=0.9 or 90%
X.drop(['amnt_loans30','daily_decr90','medianamnt_loans90','rental90'],axis=1,inplace=True)
```

#### Implementing Train_Test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 66)
```

### Applying KNN Classifier

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=25)
knn.fit(X_train,y_train)
```

```python
Knn_pred= knn.predict(X_test)
```

```python
#confussion matrix
cm= confusion_matrix(y_test,Knn_pred)
cm
```

```python
accuracy=accuracy_score(y_test,Knn_pred)
recall=recall_score(y_test,Knn_pred)
precision=precision_score(y_test,Knn_pred)
F1=f1_score(y_test,Knn_pred)
print("accuracy= {} \n recall= {} \n  preision= {} \n  f1_score= {}".format(accuracy,recall,precision,F1))
```
